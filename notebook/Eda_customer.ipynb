```python
numerical_col=df.select_dtypes(include="number")
categorical_col=df.select_dtypes(include="O")
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
sns.histplot(df['Age'],kde=True)
plt.title("age")

plt.subplot(1,3,2)
sns.histplot(df["Annual Income (k$)"],kde=True)
plt.title("annual income")

plt.subplot(1,3,3)
sns.histplot(df["Spending Score (1-100)"],kde=True)
plt.title("Spending Score")
plt.show()
```

```python
 plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
sns.boxplot(df['Age'],orient="h")
plt.title("age")

plt.subplot(1,3,2)
sns.boxplot(df["Annual Income (k$)"],orient="h")
plt.title("annual income")

plt.subplot(1,3,3)
sns.boxplot(df["Spending Score (1-100)"],orient="h")
plt.title("Spending Score")
plt.show()
```

```python
sns.countplot(data=df,x='Gender')
plt.title('Gender')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='pie',autopct="%0.1f%%")
plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.show()            
```

```python
for c_c in categorical_col:
    for n_c in numerical_col:
        sns.boxplot(x=c_c,y=n_c,data=df)
        plt.show()
```

```python
for c_c in categorical_col:
    for n_c in numerical_col:
        sns.barplot(x=c_c,y=n_c,data=df,estimator='mean')
        plt.show()
```

```python
for c_c in categorical_col:
    for n_c in numerical_col:
        sns.stripplot(x=c_c,y=n_c,data=df,jitter=True)
        plt.show()
```
